```python
from pathlib import Path
import sys
sys.path.append("..")
from data_setup import SHHSDataset
```

```python
dir = Path("/media/nvme1/sleep/ECG/DREAMT")
dataset = SHHSDataset(dir, "dreamt_ecg", multiplier=1e6, downsample_rate=4, return_file_name=True)
```

```python
wake_threshold = 0.5
rem_threshoald = 0.05

subjects_with_high_wake = []
subjects_with_low_rem = []
for idx in range(len(dataset)):
    _, labels, _, filename = dataset[idx]

    # Count the occurrences of each label
    count_wake = (labels == 0).sum().item()
    count_nrem = (labels == 1).sum().item()
    count_rem = (labels == 2).sum().item()
    total_count = len(labels)

    wake_percentage = count_wake / total_count
    rem_percentage = count_rem / total_count

    if wake_percentage > wake_threshold:
        subjects_with_high_wake.append(filename)
    
    if rem_percentage < rem_threshoald:
        subjects_with_low_rem.append(filename)
```

```python
print(f"Number of subjects with more than {wake_threshold * 100:.0f}% wake epochs: {len(subjects_with_high_wake)}")
print(f"Number of subjects with less than {rem_threshoald * 100:.0f}% of REM sleep: {len(subjects_with_low_rem)}")
```
