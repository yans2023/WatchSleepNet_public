```python
import numpy as np
import matplotlib.pyplot as plt

# Data: mean values and standard deviations
metrics = ["Accuracy", "F1", "REM F1", "AUROC", "Kappa"]
models = ["Finetuned on TCN+LSTM+Attention+Classifier", "Finetuned on All Layers"]

# Representative Sampling means and standard deviations
rep_means = np.array(
    [[0.7630, 0.7559, 0.3777, 0.8808, 0.5601], [0.7743, 0.7678, 0.4013, 0.8880, 0.5816]]
)

rep_stds = np.array(
    [[0.0073, 0.0077, 0.0131, 0.0037, 0.0126], [0.0087, 0.0091, 0.0286, 0.0072, 0.0173]]
)

# Selective Sampling means and standard deviations
sel_means = np.array(
    [[0.7182, 0.7134, 0.4091, 0.8363, 0.4794], [0.7401, 0.7376, 0.4287, 0.8503, 0.5240]]
)

sel_stds = np.array(
    [[0.0077, 0.0076, 0.0049, 0.0034, 0.0138], [0.0053, 0.0058, 0.0086, 0.0029, 0.0104]]
)

# Plot configuration
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), sharex=True)
bar_width = 0.2  # width of the bars (even narrower)
index = np.arange(len(metrics))  # locations for the groups

# Colors for the bars
rep_color = "lightsteelblue"
sel_color = "royalblue"

# Larger font sizes
font_size = 14

# Plot for TCN+LSTM+Attention+Classifier
ax1.bar(
    index,
    rep_means[0],
    bar_width,
    yerr=rep_stds[0],
    label="Representative Sampling",
    color=rep_color,
    capsize=6,
    error_kw={"elinewidth": 2},
)
ax1.bar(
    index + bar_width,
    sel_means[0],
    bar_width,
    yerr=sel_stds[0],
    label="Selective Sampling",
    color=sel_color,
    capsize=6,
    error_kw={"elinewidth": 2},
)
ax1.set_title(models[0], fontsize=font_size + 2)
ax1.set_ylabel("Scores", fontsize=font_size)
ax1.legend(fontsize=font_size)

# Plot for All Layers
ax2.bar(
    index,
    rep_means[1],
    bar_width,
    yerr=rep_stds[1],
    label="Representative Sampling",
    color=rep_color,
    capsize=6,
    error_kw={"elinewidth": 2},
)
ax2.bar(
    index + bar_width,
    sel_means[1],
    bar_width,
    yerr=sel_stds[1],
    label="Selective Sampling",
    color=sel_color,
    capsize=6,
    error_kw={"elinewidth": 2},
)
ax2.set_title(models[1], fontsize=font_size + 2)
ax2.set_ylabel("Scores", fontsize=font_size)

# Set ticks and labels
ax2.set_xticks(index + bar_width / 2)
ax2.set_xticklabels(metrics, fontsize=font_size)
ax1.tick_params(axis="both", labelsize=font_size)
ax2.tick_params(axis="both", labelsize=font_size)

# Format y-axis tick labels to two decimal places
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x:.2f}"))
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x:.2f}"))

# Layout
plt.tight_layout()
plt.show()
```
